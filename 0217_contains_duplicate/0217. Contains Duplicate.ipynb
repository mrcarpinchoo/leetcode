**217. Contains Duplicate**: Given an integer array `nums`, return `true` if any value appears at least twice in the array, and return `false` if every element is distinct.

**Example 1**:

Input: `nums = [1,2,3,1]`

Output: `true`

Explanation: The element 1 occurs at the indices 0 and 3.

**Example 2**:

Input: `nums = [1,2,3,4]`

Output: `false`

Explanation: All elements are distinct.

**Example 3**:

Input: `nums = [1,1,1,3,3,4,3,2,4,2]`

Output: `true`

**Constraints**:
- `1 <= nums.length <= 105`
- `109 <= nums[i] <= 109`

**Topics**: Array, Hash Table, Sorting

### Solution 1: Hash Table Approach
This solution determines whether the array contains any duplicate values by using a hash table to track which numbers have already been seen.

1. Create an empty hash table.
2. Iterate through each number in `nums`.
    - If the number is already in the hash table, return `True` (a duplicate was found).
    - Otherwise, insert the number into the hash table.
3. If the loop finishes without finding duplicates, return `False`.

The key idea is to use the hash table for constant-time membership checking. As soon as a number appears more than once, the algorithm immediately detects it and stops.

- Time complexity: $O(n)$
- Space complexity: $O(n)$

```python
from typing import List

class Solution:
    def containsDuplicate(self, nums: List[int]) -> bool:
        hash_table = {}

        for num in nums:
            if num in hash_table: return True
            else: hash_table[num] = num
            # end if-else
        # end for-in

        return False
    # end def
# end class
```

```python
solution = Solution()

print(solution.containsDuplicate([1, 2, 3, 1])) # True
print(solution.containsDuplicate([1, 2, 3, 4])) # False
print(solution.containsDuplicate([1, 1, 1, 3, 3, 4, 3, 2, 4, 2])) # True
```

### Solution 2: Set Approach
This solution checks whether the array contains duplicates by using a set to store the elements that have already been seen.

1. Create an empty set called seen.
2. Iterate through each number in nums.
    - If the number is already in seen, return True (a duplicate exists).
    - Otherwise, add the number to the set.
3. If the loop completes without finding duplicates, return False.

The key idea is that a set allows constant-time membership checking on average. As soon as a repeated element appears, the algorithm immediately detects it and stops.

- Time complexity: $O(n)$
- Space complexity: $O(n)$

```python
from typing import List

class Solution:
    def containsDuplicate(self, nums: List[int]) -> bool:
        seen = set()

        for num in nums:
            if num in seen: return True
            
            seen.add(num)
        # end for-in

        return False
    # end def
# end class
```

```python
solution = Solution()

print(solution.containsDuplicate([1, 2, 3, 1])) # True
print(solution.containsDuplicate([1, 2, 3, 4])) # False
print(solution.containsDuplicate([1, 1, 1, 3, 3, 4, 3, 2, 4, 2])) # True
```
